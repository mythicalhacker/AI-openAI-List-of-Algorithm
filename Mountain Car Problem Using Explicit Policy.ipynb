```python
import gym
import numpy as np
import matplotlib.pyplot as plt
```

```python
def explicitPolicy(velocity):
    if velocity<0:
        return 0
    elif velocity>=0:
        return 2
```

```python
if __name__=='__main__':
    env=gym.make('MountainCar-v0')
    num=100
    reward=np.zeros(num)
    for i in range(num):
        obs=env.reset()
        ep=0
        finished=False
        while not finished:
            action=explicitPolicy(obs[1])
            obs,r,finished,info=env.step(action)
            ep +=r
        reward[i]=ep
    plt.plot(reward)
    plt.show()
```
